```python
def calc_saida_pedido(horas, minutos, tempo_pedido):
    pedido_horas = tempo_pedido // 60
    pedido_minutos = tempo_pedido % 60

    saida_horas = horas + pedido_horas
    saida_minutos = minutos + pedido_minutos

    if saida_minutos >= 60:
      saida_horas = saida_horas + 1
      saida_minutos = saida_minutos - 60

    return saida_horas, saida_minutos

pedidos = 0
leu_primeiro = False

while True:
    horas, minutos, tempo_pedido = [int(x) for x in input().replace(":", " ").split()]
    if horas == 0 and minutos == 0:
      break

    if leu_primeiro:
      if (horas < saida_horas) or ( (horas == saida_horas) and (minutos < saida_minutos) ):
        horas, minutos = saida_horas, saida_minutos
    else:
      leu_primeiro = True

    saida_horas, saida_minutos = calc_saida_pedido(horas, minutos, tempo_pedido)

    if saida_horas < 23:
      pedidos = pedidos + 1
    else:
      break

print(pedidos)
```
