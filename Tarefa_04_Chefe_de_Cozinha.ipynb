```python
path = '/content/open/'

#-------------------------------------------------------------------------------
# YOU NEED THIS
#
def calc_saida_pedido(horas, minutos, tempo_pedido):
    tempo_pedido_horas = tempo_pedido // 60
    tempo_pedido_minutos = tempo_pedido % 60

    tempo_total_horas = horas + tempo_pedido_horas
    tempo_total_minutos = minutos + tempo_pedido_minutos

    if tempo_total_minutos >= 60:
      tempo_total_horas += 1
      tempo_total_minutos -= 60

    return tempo_total_horas, tempo_total_minutos
#
#-------------------------------------------------------------------------------

for exercicio in range(1, 10):
  exercicio_str = f"{exercicio:02}"
  print(f"\nExerc√≠cio {exercicio_str}")

  pedidos = 0
  tempo_total_horas = 0
  tempo_total_minutos = 0

#-------------------------------------------------------------------------------
# YOU NEED THIS
#

  with open(path + 'arq' + exercicio_str + '.in', 'r') as f:
    for line in f:

      # input
      horas, minutos, tempo_pedido = [int(x) for x in line.replace(":", " ").split()]
      # print(horas, minutos, tempo_pedido)

      # se o pedido est√° sendo feito antes, o hor√°rio de in√≠cio √© substitu√≠do pelo t√©rmino do pedido anterior
      if tempo_total_horas != 0:
       if (horas < tempo_total_horas) or ( (horas == tempo_total_horas) and (minutos < tempo_total_minutos) ):
          horas, minutos = tempo_total_horas, tempo_total_minutos

      # calcula hora de sa√≠da do pedido
      tempo_total_horas, tempo_total_minutos = calc_saida_pedido(horas, minutos, tempo_pedido)
      # print(f"{tempo_total_horas:02}:{tempo_total_minutos:02}")

      # verifica se o pedido pode ser aceito, isto √© < 23h (n√£o aceita igual!)
      if tempo_total_horas < 23: # or (tempo_total_horas == 23 and tempo_total_minutos == 0):
        # print("Pedido aceito")
        pedidos = pedidos + 1
      else:
        None
        break
        # print("Pedido rejeitado")

  print(pedidos)
#
#-------------------------------------------------------------------------------

  print(f"Total de pedidos aceitos: {pedidos}")

  with open(path + 'arq' + exercicio_str + '.out', 'r') as f:
    for line in f:
      pedidos_gabarito = int(line.replace('\n',''))
      print(f"Gabarito:                 {pedidos_gabarito}")
      if pedidos_gabarito == pedidos:
        print("Resposta correta")
      else:
        print("Resposta incorreta")

```
