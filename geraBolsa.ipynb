```
# Import the yfinance. If you get module not found error the run !pip install yfinance from your Jupyter notebook
import yfinance as yf

# Get the data for the stock AAPL
data = yf.download('ITUB3.SA','2019-01-01','2021-09-10')

# Import the plotting library
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the close price of the AAPL
data['Adj Close'].plot()
plt.show()
```

```
# Define the ticker list
import pandas as pd
# tickers_list = ['^BVSP', 'ITUB3.SA', 'ITUB4.SA', 'BRL=X']
tickers_list = ['^BVSP', 'ITUB3.SA', 'ITUB4.SA']

# Fetch the data
import yfinance as yf
data = yf.download(tickers_list,'2021-06-01','2021-09-10')['Adj Close']

# Print first 5 rows of the data
print(data.head())
```

```
data.head()
```

```
df = data.reset_index()
df = df[['Date','ITUB3.SA','ITUB4.SA']]
df.head()
df = df.melt(id_vars='Date')
df.head()
```

Source: https://towardsdatascience.com/historical-stock-price-data-in-python-a0b6dc826836

```
import plotly.express as px

fig = px.line(df, x='Date', y='value', color='variable')
fig.show()
```

```
# Plot all the close prices
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Returns", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Cumulative Returns', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
# Plot all the close prices
(data[['ITUB3.SA','ITUB4.SA']].plot(figsize=(10, 7)))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Values", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Values', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```

```
data['DIFF'] = data['ITUB4.SA'] - data['ITUB3.SA']
```

```
# Plot all the close prices
(data[['DIFF']].plot(figsize=(10, 7)))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Diferena em Reais ITUB4 - ITUB3", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Values', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
