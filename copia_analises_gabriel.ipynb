```python
import dask.dataframe as dd
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from dask.diagnostics import ProgressBar
from tqdm import tqdm

# Função para validar se uma string parece ser uma data no formato esperado
def is_valid_date(date_str):
    if isinstance(date_str, str):
        return bool(re.match(r"^\d{1,2}/\d{1,2}/\d{4}$", date_str))
    return False

# Função para tentar converter o valor em uma data
def parse_date(date_str):
    if is_valid_date(date_str):
        try:
            # Tentar converter no formato DD/MM/YYYY
            return pd.to_datetime(date_str, format='%d/%m/%Y', errors='coerce')
        except ValueError:
            # Se falhar, tentar no formato MM/DD/YYYY
            return pd.to_datetime(date_str, format='%m/%d/%Y', errors='coerce')
    else:
        return pd.NaT  # Se não for uma data válida

# Função para categorizar a pista de partida
def categorize_airport_type(departure):
    if pd.isna(departure):
        return pd.NA  # Retornar valor ausente se o dado estiver faltando
    return 'Curta' if departure in pista_curta else 'Longa'

# Carregar os dados de remoção com dask
path_raw = ""
df_remocoes = dd.read_csv(path_raw + 'remocao_componentes_sheet1.csv', assume_missing=True)

# Aplicar a função de conversão de data com o `map_partitions`
df_remocoes['Rem. Date'] = df_remocoes['Rem. Date'].map_partitions(lambda df: df.apply(parse_date))

# Filtrar apenas as remoções válidas
df_remocoes_validas = df_remocoes.dropna(subset=['Rem. Date'])

# Carregar os dados de voos com dask
path = ""
df_voos = dd.concat([dd.read_csv(path + 'VOOS MAX_0.csv'),
                     dd.read_csv(path + 'VOOS NG PART 1_0.csv'),
                     dd.read_csv(path + 'VOOS NG PART 2_0.csv'),
                     dd.read_csv(path + 'VOOS NG PART 3_0.csv')])

# Converter a coluna de data
df_voos['departure_date'] = dd.to_datetime(df_voos['departure_date'], dayfirst=True)

# Lista de aeroportos de pista curta
pista_curta = ['SDU', 'IOS', 'NVT', 'CXJ', 'CPV', 'CGH', 'JOI', 'UDI', 'IZA']

# Categorizar a pista de partida usando uma função separada
df_voos['departure_airport_type'] = df_voos['departure'].map_partitions(lambda df: df.apply(categorize_airport_type))

# Combinar os dados de voos e remoções
df_combined = dd.merge(df_remocoes_validas, df_voos, left_on=['AC Reg.', 'Rem. Date'], right_on=['ac_registr', 'departure_date'], how='inner')

# Ativar o progresso com o tqdm e dask
pbar = tqdm(total=len(df_remocoes_validas))
with ProgressBar():
    df_combined = df_combined.compute()
    pbar.update()

```

```python
# Criar uma coluna de Remoção Planejada (1 para removido, 0 para não removido)
df_combined['Removal Status'] = df_combined['Removal Type'].apply(lambda x: 1 if x in ['US', 'SD'] else 0)

# Análise Estatística: Teste Qui-Quadrado
contingency_table = pd.crosstab(df_combined['departure_airport_type'], df_combined['Removal Status'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Tabela de Contingência:\n", contingency_table)
print(f"Qui-quadrado: {chi2}")
print(f"p-valor: {p}")
print(f"Graus de Liberdade: {dof}")

if p < 0.05:
    print("Resultado: Existe uma correlação estatisticamente significativa entre o tipo de pista e a remoção de componentes.")
else:
    print("Resultado: Não existe uma correlação estatisticamente significativa entre o tipo de pista e a remoção de componentes.")

# Visualização: Gráfico de Barras (Contagem de Remoções por Tipo de Pista e Tipo de Remoção)
sns.countplot(data=df_combined, x='departure_airport_type', hue='Removal Status')
plt.title('Distribuição de Remoções por Tipo de Pista')
plt.xlabel('Tipo de Pista de Partida')
plt.ylabel('Número de Remoções')
plt.show()

# Contagem de Remoções por Tipo de Pista
planned_removals = df_combined[df_combined['Removal Status'] == 1]['departure_airport_type'].value_counts()

print(f'Total de remoções em pistas curtas:\n{planned_removals}')
```

# Regressão logistica

#### Vamos usar a regressão logística para prever a probabilidade de remoção de um componente (Removal Status) com base no tipo de pista e outras variáveis explicativas.

### Explicação

####    Variáveis Explicativas:
  *  departure_airport_type_encoded (1 para pistas curtas, 0 para pistas longas) e TSN (tempo desde novo).
####    Variável Dependente:
  *  Removal Status (0 para não removido, 1 para removido).

####    Interpretação:
  *  O coeficiente de cada variável mostra o impacto relativo na probabilidade de remoção. Um coeficiente positivo aumenta a chance de remoção, enquanto um negativo diminui.

```python
import statsmodels.api as sm
import pandas as pd

# Preparação dos dados para regressão logística
# Usaremos 'departure_airport_type' e 'TSN' (Tempo desde novo) como variáveis explicativas
df_combined['departure_airport_type_encoded'] = df_combined['departure_airport_type'].apply(lambda x: 1 if x == 'Curta' else 0)

# Definindo as variáveis independentes e a variável dependente
X = df_combined[['departure_airport_type_encoded', 'TSN']]  # Variáveis explicativas
y = df_combined['Removal Status']  # Variável dependente

# Adiciona a constante ao modelo
X = sm.add_constant(X)

# Ajustando o modelo de regressão logística
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Sumário do modelo
print(result.summary())
```

# Modelo de Regressão de Cox (Modelo de Risco Proporcional de Cox)

#### Aqui, aplicamos o modelo de Cox para modelar o tempo até a remoção de um componente com base no tipo de pista e outras variáveis.

### Explicação

####    Variáveis Explicativas:
  *  duration (tempo até o evento), que é o tempo desde novo (TSN).
####    Variável Dependente:
  *  event (0 para não removido, 1 para removido).

####    Interpretação:
  *  O modelo de Cox estima o risco proporcional de remoção de um componente ao longo do tempo, permitindo avaliar o efeito das variáveis explicativas sobre o tempo de remoção.

```python
!pip install lifelines
```

```python
from lifelines import CoxPHFitter

# Preparação dos dados para o modelo de Cox
# Usamos 'TSN' como o tempo até o evento e 'Removal Status' como o indicador de evento
df_combined['event'] = df_combined['Removal Status']
df_combined['duration'] = df_combined['TSN']

# Selecionando as variáveis para o modelo
df_cox = df_combined[['duration', 'event', 'departure_airport_type_encoded', 'TSN']]

# Ajustando o modelo de Cox
cph = CoxPHFitter()
cph.fit(df_cox, duration_col='duration', event_col='event')

# Exibindo o sumário do modelo de Cox
cph.print_summary()

# Plotando os coeficientes do modelo de Cox
cph.plot()
plt.title("Coeficientes do Modelo de Cox")
plt.show()
```
