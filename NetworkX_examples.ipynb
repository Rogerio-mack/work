```python
# prompt: Crie agora rede watts_strogatz_graph, barabasi_albert_graph, random_lobster e erdos_renyi_graph, todas de 100 nós cada. Exiba as redes e suas principais estatísticas.

import networkx as nx
import matplotlib.pyplot as plt

# Cria uma rede Watts-Strogatz com 100 nós, vizinhos próximos k=4 e probabilidade de reconexão p=0.1
watts_strogatz = nx.watts_strogatz_graph(100, 4, 0.1)

# Cria uma rede Barabási-Albert com 100 nós e m=2 novas conexões por nó
barabasi_albert = nx.barabasi_albert_graph(100, 2)

# Cria uma rede random_lobster com 100 nós
random_lobster = nx.random_lobster(100, 0.1, 0.3)

# Cria uma rede Erdős–Rényi com 100 nós e probabilidade de conexão p=0.1
erdos_renyi = nx.erdos_renyi_graph(100, 0.1)

def plot_graph_stats(graph, title):
  plt.figure(figsize=(4,4))
  nx.draw(graph, node_size=50, node_color="skyblue", edge_color="gray", width=0.5)
  plt.title(title)
  plt.show()

  print(f"--- {title} ---")
  print("Número de nós:", graph.number_of_nodes())
  print("Número de arestas:", graph.number_of_edges())
  print("Densidade da rede:", nx.density(graph))
  try:
    print("Diâmetro da rede:", nx.diameter(graph))
  except nx.NetworkXError:
    print("Diâmetro da rede: Não definido (rede desconexa)")
  print("Coeficiente de agrupamento médio:", nx.average_clustering(graph))

  print()

plot_graph_stats(watts_strogatz, "Rede Watts-Strogatz (100 nós)")
plot_graph_stats(barabasi_albert, "Rede Barabási-Albert (100 nós)")
plot_graph_stats(random_lobster, "Rede Random Lobster (100 nós)")
plot_graph_stats(erdos_renyi, "Rede Erdos-Renyi (100 nós)")
```
