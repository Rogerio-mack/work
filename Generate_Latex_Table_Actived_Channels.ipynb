```python
import pandas as pd
from IPython.display import display, Latex, HTML

pd.set_option('display.float_format', lambda x: '%.2f' % x)

```

```python
df = pd.read_excel('/content/Active_ttest_6MBB_16Bebes.xlsx',sheet_name='Actived M')
df
```

```python
def gera_latex_table(df, titulo):

  # warnings.filterwarnings('ignore')
  # warnings.filterwarnings('default')

#  df.style.clear()
#  df.style.table_styles = []
#  df.style.caption = None
#  df.style.format({
#   ("Numeric", "Integers"): '\${}',
#   ("Numeric", "Floats"): '{:.3f}',
#   ("Non-Numeric", "Strings"): str.upper
#  })

#  (df.style.format(precision=2)
#     .hide(level=0, axis=0))

#  latex_table = df.style.to_latex(hrules=True)

  latex_table = df.to_latex(index=None,float_format="%.3f")
  # display(Latex(latex_table))

  latex_table = latex_table.replace('tabular','array')
  latex_table = latex_table.replace('toprule','hline')
  latex_table = latex_table.replace('midrule','hline')
  latex_table = latex_table.replace('bottomrule','hline')

  latex_table = '\\textbf{' + titulo + '} \\\\' + latex_table


  return latex_table
```

```python
# Display the LaTeX code as a LaTeX document
# display(Latex('\\textbf{dCohen-test values} \\\\'))
display(Latex(gera_latex_table(df,'HBO Mother - All Channels')))
display('')
display(Latex(gera_latex_table(df[df.ACTIVED == True],'HBO Mother - Only Actived Channels')))
# Save the LaTeX code to a file
with open('output.tex', 'w') as f:
    f.write(gera_latex_table(df,'HBO Mother - Only Actived Channels'))

```
