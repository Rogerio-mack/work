```python
#@title **Equação do Plano (somente equação)** 
#@markdown Informe **3 pontos** no formato **(x,y,z)** como nos exemplos e click no botão para executar.
import numpy as np
from IPython.display import display, Math

a = '(1,2,3)' #@param {type:"string"}
b = '(-1,0,1)' #@param {type:"string"}
c = '(1,1,1)' #@param {type:"string"}

def plan(a,b,c):
  pA = np.array(eval(a))
  pB = np.array(eval(b))
  pC = np.array(eval(c))

  AB = pB - pA
  AC = pC - pA

  N = np.cross(AB,AC)
  d = - N.dot(pA)

  return pA, pB, pC, N, d

pA, pB, pC, N, d = plan(a,b,c) 

display(Math(r'{:1.2f}x + {:1.2f}y + {:1.2f}z + {:1.2f} = 0'.format(N[0],N[1],N[2],d)))

 
```

```python
#@title **Equação do Plano (equação e gráfico)** 
#@markdown Informe **3 pontos** no formato **(x,y,z)** como nos exemplos e click no botão para executar.
from IPython.display import display, Math

a = '(1,2,3)' #@param {type:"string"}
b = '(-1,0,1)' #@param {type:"string"}
c = '(1,1,1)' #@param {type:"string"}

def plan(a,b,c):
  pA = np.array(eval(a))
  pB = np.array(eval(b))
  pC = np.array(eval(c))

  AB = pB - pA
  AC = pC - pA

  N = np.cross(AB,AC)
  d = - N.dot(pA)

  return pA, pB, pC, N, d

def plan_plot(pA, pB, pC, N, d):
  import matplotlib.pyplot as plt
  from mpl_toolkits.mplot3d import Axes3D

  xx, yy = np.meshgrid(range(-10,10), range(-10,10))

  z = (-N[0] * xx -N[1] * yy - d) * 1. / N[2]

  ax = plt.figure(figsize=(9,7)).gca(projection='3d')

  # ax.quiver(pA[0],pA[1],pA[2],N[0],N[1],N[2],color='r',
  #          pivot='tail',length=20,arrow_length_ratio=0.3/2)
  
  ax.plot_surface(xx, yy, z, alpha=0.3)

  ax.set_title('Plano',size=14, fontweight="bold")

  # ax.set_xlim([-10,10])
  # ax.set_ylim([-10,10])
  # ax.set_zlim([-10,10])
  ax.set_xlabel("X", size=12, fontweight="bold")
  ax.set_ylabel("Y", size=12, fontweight="bold")
  ax.set_zlabel("Z", size=12, fontweight="bold")
  plt.show()

  return

pA, pB, pC, N, d = plan(a, b, c)  

display(Math(r'{:1.2f}x + {:1.2f}y + {:1.2f}z + {:1.2f} = 0'.format(N[0],N[1],N[2],d)))

plan_plot(pA, pB, pC, N, d)

```
