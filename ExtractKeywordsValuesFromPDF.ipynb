# Extract numbers from selected lines

```python
import re

values = []
for line in selected_lines:
  s = line[ len(keyword)::] # skip first keyword characters
  rr = re.findall("[\d]+[\,]?\d", s)
  values.append(rr[0])

print(values)  
```
